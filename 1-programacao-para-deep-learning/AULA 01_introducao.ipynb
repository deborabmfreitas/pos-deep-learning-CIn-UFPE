## Variáveis numéricas - conversão de tipos

Se os dois operandos x e y forem inteiros, o resultado é um
valor inteiro (a não ser que a operação seja divisão).
Caso haja pelo menos um operando de representação
numérica maior, o resultado será no tipo de maior
representação.

```python
# Exemplo 1

x = 862
y = 4
z = x + y
type(z)
```

```python
# Exemplo 2

x = 862.0
y = 4
z = x + y
type(z)
```

## Variáveis de tipo boolean (tipo lógico)

Armazenam valores lógicos: Verdadeiro (True) ou Falso (False).

| Operação | Formato        | Funcionamento                                                   | Exemplo                                                                 |
|----------|----------------|------------------------------------------------------------------|-------------------------------------------------------------------------|
| E (and)  | var1 and var2  | `True and True = True`<br>`True and False = False`<br>`False and True = False`<br>`False and False = False` | x = True<br>y = False<br>z = x and y                                   |
| Ou (or)  | var1 or var2   | `True or True = True`<br>`True or False = True`<br>`False or True = True`<br>`False or False = False` | a = False<br>b = True<br>c = a or b                                    |
| Não (not)| not var1       | `not True = False`<br>`not False = True`                       | ehCasado = False<br>ehCasadoDepois = not ehCasado                      |


```python
# Exemplo 1

a = True
b= False
type(a), type(b)
```

```python
# Exemplo 2

a = 2
b = 5
a > b
```

O retorno de operações de comparação sempre é um valor lógico Verdadeiro ou Falso.

| Operação               | Formato       | Exemplo                         |
|------------------------|---------------|---------------------------------|
| Maior que              | var1 > var2   | x = 10<br>y = 5<br>z = x > y    |
| Maior ou igual que     | var1 >= var2  | x = 10<br>y = 10<br>z = x >= y  |
| Menor que              | var1 < var2   | x = 3<br>y = 7<br>z = x < y     |
| Menor ou igual que     | var1 <= var2  | x = 5<br>y = 5<br>z = x <= y    |
| Igual                  | var1 == var2  | x = 8<br>y = 8<br>z = x == y    |

### Convertendo variáveis de tipo explicitamente

- É possível converter uma variável de um determinado tipo para outro válido
- Isso é possível se a variável obedecer a algumas condições

| Conversão                               | Formato do Construtor | Exemplo                                                                 | Alertas                                                                                   |
|----------------------------------------|------------------------|-------------------------------------------------------------------------|--------------------------------------------------------------------------------------------|
| De inteiro para float                  | float(var1)            | x = 2<br>y = float(x)                                                   | Precisa ser compatível com a representação de ponto flutuante da máquina.                |
| De float para inteiro                  | int(var1)              | x = 20.34<br>y = int(x)                                                 | Arredonda **para baixo**, descartando as casas decimais.                                  |
| De string para inteiro                 | int(var1)              | x = "6412"<br>y = int(x)                                                | O valor precisa ser uma string representando um número inteiro válido.                   |
| De string para float                   | float(var1)            | x = "223.42"<br>y = float(x)                                            | O valor precisa ser uma string representando um número real válido.                      |
| De inteiro/float para string           | str(var1)              | x = 346.21<br>y = str(x)                                                | Converte o valor numérico em uma string.                                                 |
| De lógico (bool) para inteiro          | int(var1)              | x = True<br>y = int(x)<br>z = False<br>w = int(z)                       | `True` se torna `1` e `False` se torna `0`.                                               |
| De lógico (bool) para float            | float(var1)            | x = True<br>y = float(x)<br>z = False<br>w = float(z)                   | `True` se torna `1.0` e `False` se torna `0.0`.                                           |
| De int ou float para lógico (bool)     | bool(var1)             | x = 23<br>y = bool(x)<br>z = 0.0<br>w = bool(z)                         | Qualquer valor diferente de zero vira `True`; `0` ou `0.0` vira `False`.                 |
