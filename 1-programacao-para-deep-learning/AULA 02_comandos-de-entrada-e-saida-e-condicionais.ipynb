# Comandos condicionais

* Nas decisões, têm-se uma condição que está sendo analisada
* `if`, `else`, `elif`

Sintaxe:

if < condicao >:  
 > comandos a serem executados caso a condição seja verdadeira  

else:  
 > comandos a serem executados caso a condição seja falsa

* Expressão booleana: o resultado é sempre True ou False
* Indentação: permite identificar quais comandos estão dentro do if e do else

```python
# Exemplo

idade = int ( input('Digite sua idade: ') )
if idade >=18:
    print('você é maior de idade')
else:
    print('você não é maior de idade')
```

Podemos ter vários comandos `elif` (else if), assim como comandos `if` e `else` dentro de if e else.
