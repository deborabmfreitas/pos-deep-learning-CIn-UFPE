# Comandos de iteração

- Muita das vezes, precisamos repetir um conjunto de ações enquanto uma
condição for satisfeita
- Essas repetições permitem o programa executar instruções sobre algum critério
de parada
- Nosso programa em Python sempre é
executado linha após linha, de cima para
baixo
- Os comandos condicionais permitem que
executemos um conjunto de comandos
em detrimento de outros

`while`e `for`


while < condicao >:
> comandos a serem executados caso a condição seja verdadeira

## Mais comandos

`break`: permite encerrar a execução de um laço no meio de sua execução

`continue`: permite passar para a próxima iteração no meio do laço. OBS: Dificulta a compreensão do código, pois há maisde uma condição a ser verificada, mas é útil quando precisamos pular os próximos comandos no meio do bloco, indo direto paraa próxima repetição

`pass`: é utilizada em instruções de controle de execução ou em funções quando a sintaxe do Python exige que se tenha um corpo. A instrução pass é útil também como marcador de lugar, quando um corpo de código ainda não tiver sido implementado
