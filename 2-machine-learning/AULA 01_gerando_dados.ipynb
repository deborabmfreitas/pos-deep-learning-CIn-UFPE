# Organização dos dados

```python
import pandas as pd
import numpy as np
```

## Dados de classificação

```python
# Dados históricos

dic_classificacao = {
    'Pessoa': ['A', 'B', 'C', 'D', 'E', 'F', 'G'],
    'Altura (m)': [1.70, 1.65, 1.80, 1.75, 1.60, 1.85, 1.68],
    'Peso (kg)': [70, 90, 80, 95, 50, 110, 68],
    'Classe': [
        'Atleta',      # A: IMC ~24.22
        'Não Atleta',  # B: IMC ~33.06
        'Atleta',      # C: IMC ~24.69
        'Não Atleta',  # D: IMC ~31.02
        'Atleta',      # E: IMC ~19.53
        'Não Atleta',  # F: IMC ~32.15
        'Atleta'       # G: IMC ~24.07
    ]
}

df_historicos = pd.DataFrame(dic_classificacao)


# Novo exemplo

dic_novo_exemplo = {
    'Pessoa':['H'],
    'Altura (m)':[1.72],
    'Peso (kg)': [72],
    'Classe': [np.nan]
}

df_novo = pd.DataFrame(dic_novo_exemplo)
```

```python
# Históricos

df_historicos
```

```python
# Dados novos

df_novo
```

```python
# Salvando

df_historicos.to_csv('data/df_historicos_classificacao',index=False)
df_novo.to_csv('data/df_novo_classificacao',index=False)
```

## Dados de regressão

```python
# Dados históricos

dic_regressao = {
    'Pessoa': ['A', 'B', 'C', 'D', 'E', 'F', 'G'],
    'Altura (m)': [1.70, 1.65, 1.80, 1.75, 1.60, 1.85, 1.68],
    'Peso (kg)': [70, 90, 80, 95, 50, 110, 68],
    'TempoCorrida(min)': [
        28.0,  # A - corredor recreativo
        45.5,  # B - corredor lento
        22.0,  # C - corredor intermediário
        55.0,  # D - muito lento (caminhada/corrida leve)
        20.0,  # E - corredor rápido
        50.0,  # F - lento
        25.5   # G - corredor intermediário
    ]
}

df_historicos = pd.DataFrame(dic_regressao)


# Novo exemplo

dic_novo_exemplo = {
    'Pessoa': ['H'],
    'Altura (m)': [1.72],
    'Peso (kg)': [72],
    'TempoCorrida(min)': [np.nan]  # valor desconhecido a ser previsto
}

df_novo = pd.DataFrame(dic_novo_exemplo)
```

```python
# Históricos

df_historicos
```

```python
# Dados novos

df_novo
```

```python
# Salvando

df_historicos.to_csv('data/df_historicos_regressao',index=False)
df_novo.to_csv('data/df_novo_regressao',index=False)
```
